```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

y_train = [[45000],[50000],[60000],[80000],[110000],[150000],[200000],[300000],[500000],[1000000]]
x_plot = np.linspace(0,len(y_train), len(y_train))#returns evenly spaced numbers array between start=0, stop=28, number needed=26
plt.scatter(x_plot, y_train, color='navy', s=30, marker='o', label="training points")
colors = ['teal', 'yellowgreen', 'gold', 'red','green','violet','grey']
x_plot = x_plot.reshape(-1,1)

for count, degree in enumerate([1, 2, 3]):
  model = make_pipeline(PolynomialFeatures(degree), Ridge())
  model.fit(x_plot, y_train)
  y_plot = model.predict(x_plot)
  plt.plot(x_plot, y_plot, color=colors[count], linewidth=2,
           label="degree %d" % degree)

plt.legend(loc='lower right')
plt.show()
```

<h1>np.polyfit(x, y, deg)</h1>
The np.polyfit(x, y, deg) function in NumPy is used to fit a polynomial of degree deg to the data (x, y) using the method of least squares. Here's an explanation of the parameters:

x: Array-like, shape (n,)
The x-coordinates of the data points. This should be a one-dimensional array-like object, such as a list or NumPy array.
y: Array-like, shape (n,) or (n, k)
The y-coordinates of the data points. For a single set of data, y should be a one-dimensional array-like object. For multiple sets of data, y can be a two-dimensional array-like object.
deg: int
The degree of the polynomial to fit to the data. This parameter specifies the number of terms in the polynomial, including the constant term. For example, if deg is 1, the resulting polynomial will be a straight line (linear fit). If deg is 2, the resulting polynomial will be a quadratic function (a parabola), and so on.
When selecting the value of deg, consider the complexity of the relationship between x and y in your data. If the relationship appears to be linear, you might choose deg=1 for a linear fit. If the relationship seems to be more complex, you might increase the degree to capture higher-order trends in the data. However, increasing the degree of the polynomial can also lead to overfitting, where the model fits the noise in the data rather than the underlying pattern. It's essential to strike a balance between capturing the complexity of the data and avoiding overfitting.

In the example model = np.poly1d(np.polyfit(x, y, 1)), np.polyfit(x, y, 1) is fitting a first-degree polynomial (a straight line) to the data (x, y). This means that the resulting model will be a linear function.
<h1>linespace</h1>
The np.linspace() function in NumPy generates an array of evenly spaced numbers over a specified interval. It takes three parameters:

start: The starting value of the sequence.
stop: The end value of the sequence.
num: The number of evenly spaced samples to generate.
In your example, line = np.linspace(1, 10, 5), you're generating an array line containing 5 evenly spaced numbers between 1 and 10 (inclusive). Here's how each parameter affects the output:

start: The first value of the sequence. In this case, it's 1.
stop: The last value of the sequence. In this case, it's 10.
num: The number of samples to generate. In this case, it's 5.
So, the array line will contain the numbers [1.0, 3.25, 5.5, 7.75, 10.0]. Each number is evenly spaced between 1 and 10, and there are 5 numbers in total.

You can adjust these parameters based on your specific needs. For example:

To change the range of values, adjust start and stop.
To change the number of samples, adjust num.
Ensure that num is an integer greater than or equal to 2, as it represents the number of points to generate, and you need at least two points to define a line.

This code is a common practice in machine learning, specifically in supervised learning tasks. Here's what each part of the code does:

x_train, x_test, y_train, y_test: These variables are used to store the training and testing data splits. Typically, x represents the features (input variables), and y represents the target variable (output variable).
# train_test_split: This function is used to split the dataset into two subsets: the training set and the testing set. It takes the input features (x) and the target variable (y) as input parameters.
x: This parameter represents the input features (independent variables) of the dataset.
y: This parameter represents the target variable (dependent variable) of the dataset.
test_size: This parameter determines the proportion of the dataset that will be included in the testing set. In this case, test_size=0.2 means that 20% of the data will be reserved for testing, while the remaining 80% will be used for training.
random_state: This parameter is used to set a seed for the random number generator. Setting a specific value for random_state ensures that the dataset is split in the same way each time the code is run. This is useful for reproducibility, as it allows for consistent results when testing different models or parameters.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
poly=PolynomialFeatures(degree=2)
x_poly_train=poly.fit_transform(x_train)
x_poly_test=poly.transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_poly_train,y_train)
```

```python
y_pred=model.predict(x_poly_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("r2_score error =",r2_score(y_test,y_pred))
print("mean_absolute_error = ",mean_absolute_error (y_test,y_pred))
print("mean_squared_error =",mean_squared_error (y_test,y_pred))
```
