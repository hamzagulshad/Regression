```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib as mlt
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/Crop Responses.csv')
df
```

```python
x=df[['Y','PT','R','NAR','DF','DM']]
y=df[['SH']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=0)
```

```python
model= LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
plt.plot(x_test,y_predict,color='red')
plt.plot(x_test,y_test,color='blue')
```

```python
print("intercept : ",model.intercept_)
print("intercept : ",model.coef_)
```

```python
#model.predict([[1,1,1,1,113,147]])2	3	2	4	112	129
model.predict([[2,3,2,4,112,129]])
```

```python
plt.plot(x_test,y_predict,color='red')
plt.scatter(x,y,color='green')
#plt.plot(biomassy,ys,color='blue')
plt.scatter(x_test,y_test,color='blue')
```

```python
plt.plot(x_test,y_predict,color='red')
plt.plot(x_test,y_test,color='blue')
```

```python
plt.scatter(y,model.predict(x))
plt.xlabel('actual value')
plt.ylabel('predict value value')
plt.title('actual value vs predict values')
plt.show()
```

```python
plt.plot(x,y,color='blue')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("r2_score error =",r2_score(y_test,y_predict))
print("mean_absolute_error = ",mean_absolute_error (y_test,y_predict))
print("mean_squared_error =",mean_squared_error (y_test,y_predict))
```
